## 가변 전광판 데이터 다듬기

```python
df1 = pd.read_csv('서울시_가변전광판_설치현황.csv')

df1 = df1.rename_axis('id').reset_index()

df1
```

```python
# df1['설치연도'] = pd.to_numeric(df1['설치연도'])
df1['설치연도'] = df1['설치연도'].apply(lambda _: str(_))
for i in range(22):
    if len(str(i)) == 1:
        df1.loc[df1['설치연도'] == f'200{i}', '설치년도코드'] = i
    else:
        df1.loc[df1['설치연도'] == f'20{i}', '설치년도코드'] = i

df1['설치년도코드'] = df1['설치년도코드'].apply(lambda _: int(_))
        
```

```python
df1 = df1.rename(columns = {"시도명":"sido", "시군구명":"gugun", "설치년도코드":"year_code"})
```

```python
del df1['설치연도']
```

```python
df1
```

```python
df1.to_csv('가변전광판.csv', index=False, encoding='cp949')
```

## 옐로우카펫 설치 현황 데이터 다듬기

```python
df9['address'] = df9['address'].str.replace('서울특별시', '')
```
